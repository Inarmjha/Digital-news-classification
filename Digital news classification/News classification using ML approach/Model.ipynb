```python
import os
import numpy as np
import sklearn
import pandas as pd
```

```python
folder='/home/guru/Videos/Dataset-MinorProject2/data/'
```

```python
#dictinoary of labels
label_mapper={'safety.txt':0,
             'flood.txt':1,
             'strike.txt':2,
             'others.txt':3}
```

```python
li=[]
label=[]
for file_name in os.listdir(folder):
    f=open(folder+file_name,'r')
    print file_name
    for line in f.readlines():
        if len(line.split(" ")) != 1:
            li.append(line)
            label.append(label_mapper[file_name])
          #  print label_mapper[file_name]
```

```python
X=np.asarray(li)
```

```python
print X
```

```python
print X.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(X)
X_train_counts.shape
```

```python
print X_train_counts.shape
```

```python
r=X_train_counts[0,:]
```

```python
##To Visulaise the ouputs of Count Vectorizer
#print str(count_vect.get_feature_names())
print(pd.DataFrame(r.A, columns=count_vect.get_feature_names()).to_string())
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tf_transformer = TfidfTransformer(use_idf=False).fit(X_train_counts)
X_train_tf = tf_transformer.transform(X_train_counts)
```

```python
print X_train_tf
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_train_tf, label, test_size=0.33, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train,y_train)
```

```python
pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```

```python
from sklearn.linear_model import SGDClassifier
clf2=SGDClassifier(loss='hinge', penalty='l2',alpha=1e-3, random_state=42)
clf2.fit(X_train,y_train)
clf2.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf3=RandomForestClassifier()
clf3.fit(X_train,y_train)
clf3.score(X_test,y_test)
```

```python
from xgboost import XGBClassifier
clf4=XGBClassifier()
clf4.fit(X_train,y_train)
clf4.score(X_test,y_test)
```
